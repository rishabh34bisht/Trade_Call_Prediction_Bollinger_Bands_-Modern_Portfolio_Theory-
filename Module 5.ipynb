```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math as mt
%matplotlib inline
```

```python
df = pd.read_csv('DHFL.csv')
df.head()
```

```python
# calculating return %
df = df[df['Series'] == 'EQ']
df.reset_index(inplace = True, drop = True)
df['Return'] = (df['Close Price'] - df['Prev Close'])/df['Prev Close']
df.head()
```

```python
# daily return mean is the combined average of all the daily return
daily_return_mean = df['Return'].mean()
print('Daily return mean is', daily_return_mean)
```

```python
# daily return standard deviation is the combined deviation of all the daily return
daily_return_std = df['Return'].std()
print('Daily return standard deviation is ', daily_return_std )
```

```python
# annualised mean is basically average daliy return multiplied with no of trading days (i.e 252)
annualised_return_mean = daily_return_mean *252
print('Annualised mean return is ', annualised_return_mean)
```

```python
# annualised standard deviation is the product of standard deviation of daliy return with the square root of trading days 
# no of trading days = 252
annualised_return_std = daily_return_std * (mt.sqrt(252))
print('Volatility is ', annualised_return_std)
```

```python
# loading five stock to build a portfolio
df1 = pd.read_csv('HDFC.csv')
df2 = pd.read_csv('JINDALSTEL.csv')
df3 = pd.read_csv('BERGEPAINT.csv')
df4 = pd.read_csv('MUTHOOTFIN.csv')
df5 = pd.read_csv('PVR.csv')
```

```python
# calculating return % for all
df1 = df1[df1['Series'] == 'EQ']
df1.reset_index(inplace = True, drop = True)
df1['HDFC_Return'] = (df1['Close Price'] - df1['Prev Close'])/df1['Prev Close']
df2 = df2[df2['Series'] == 'EQ']
df2.reset_index(inplace = True, drop = True)
df2['JINDALSTEL_Return'] = (df2['Close Price'] - df2['Prev Close'])/df2['Prev Close']
df3 = df3[df3['Series'] == 'EQ']
df3.reset_index(inplace = True, drop = True)
df3['BERGEPAINT_Return'] = (df3['Close Price'] - df3['Prev Close'])/df3['Prev Close']
df4 = df4[df4['Series'] == 'EQ']
df4.reset_index(inplace = True, drop = True)
df4['MUTHOOTFIN_Return'] = (df4['Close Price'] - df4['Prev Close'])/df4['Prev Close']
df5 = df5[df5['Series'] == 'EQ']
df5.reset_index(inplace = True, drop = True)
df5['PVR_Return'] = (df5['Close Price'] - df5['Prev Close'])/df5['Prev Close']
df1.head()
```

```python
# creating a new data frame with return % of the 5 selected stocks 
df6 = pd.DataFrame(df1[['Date','HDFC_Return']])
df6['JINDALSTEL_Return'] = df2['JINDALSTEL_Return']
df6['BERGEPAINT_Return'] = df3['BERGEPAINT_Return']
df6['MUTHOOTFIN_Return'] = df4['MUTHOOTFIN_Return']
df6['PVR_Return'] = df5['PVR_Return'] 
df6['Date'] = df6['Date'].astype('datetime64')
df6.set_index('Date', inplace = True)
df6.head()
```

```python
# for transpose we can also use 'weights.T'
weights = np.array([.2, .2, .2, .2, .2])
weights_transpose = weights.transpose()
weights_transpose
```

```python
# np.dot() is used for matrix multiplication (dot product)
port_mean_arr = np.array([df6['HDFC_Return'].mean(), df6['JINDALSTEL_Return'].mean(), df6['BERGEPAINT_Return'].mean(), 
                          df6['MUTHOOTFIN_Return'].mean(), df6['PVR_Return'].mean()]) 

annualised_portfolio_mean_return = np.dot(port_mean_arr, weights) * 252
print('Annualised mean Portfolio return ', annualised_portfolio_mean_return )
print(port_mean_arr)
```

```python
# calculating the covariance matrix 
df_cov = df6.cov()
cov_mat = df_cov.values
cov_mat
```

```python
# Volatility can be calculated in two steps : matrix multiplication 
# it involves three matices 
# in order they are : weights matrix (1 row 5 column)
#                     covariance matrix (5 rows and 5 column)
#                     transpose of weights array (5 rows and 1 column)
# search portfolio covariance matrix
# and then for the whole year we multiplied with square root of 252
portfolio_Volatility = np.sqrt(np.dot(weights_transpose, np.dot(cov_mat, weights))) * np.sqrt(252)
print('Portfolio Volatility/Standard deviation is ',portfolio_Volatility)
```

```python
# made an array to hold results full of 1s
# it has 30,000 rows and 3 columns, for holding     1. annualised mean return
#                                                   2. portfolio volatility
#                                                   3. Sharp ratio
# with np.random.random we generated a list of 5 random numbers for weights
# to make their total sum equal to one, we divided each number by the total sum of all the random numbers
# Sharpe Ratio (return / volatility)
# we used these weights to calculate annualised mean return and volatility
# added three calculated values in the same row under different columns
# repeat the loop 30,000 times
# create a data frame of the 2D array

results = np.ones((30000,3))
for i in range(30000):
    weights = np.random.random(5)
    weights = weights / np.sum(weights)
    annualised_portfolio_mean_return = np.dot(port_mean_arr, weights) * 252
    portfolio_Volatility = np.sqrt(np.dot(weights_transpose,np.dot(cov_mat, weights))) * np.sqrt(252)
    results[i,0] = annualised_portfolio_mean_return
    results[i,1] = portfolio_Volatility
    results[i,2] = results[i,0] / results[i,1]

df_result = pd.DataFrame(results, columns=['Mean_Return','Volatility','Sharpe'])
df_result.head()
```

```python
# plotting a scatter plot for Mean return vs volatility
# the color parameter for the points is based on the values of Sharpe column
x = df_result['Volatility']
y = df_result['Mean_Return']
plt.figure(figsize = (11,7))
plt.scatter(x,y, c = df_result.Sharpe, cmap='RdYlBu')
plt.xlabel('Volatility')
plt.ylabel('Returns')
plt.colorbar()

# we obtained the row where Sharpe is max and from that assigned x and y to plot the highest sharp ratio
# we obtained the row where Volatility is lowest and from that assigned x and y to plot the lowest volatility
sharpe_ratio_highest = df_result.iloc[df_result['Sharpe'].idxmax()]
print('Highest Sharpe Ratio: \n', sharpe_ratio_highest)
Volatility_lowest = df_result.iloc[df_result['Volatility'].idxmin()]
print('Lowest Volatility: \n', Volatility_lowest)
x1 = sharpe_ratio_highest[1]
y1 = sharpe_ratio_highest[0]
x2 = Volatility_lowest[1]
y2 = Volatility_lowest[0]
plt.scatter(x1,y1, marker = '*', color = 'red', s = 250)
plt.scatter(x2,y2, marker = '*', color = 'green', s = 250)
```
