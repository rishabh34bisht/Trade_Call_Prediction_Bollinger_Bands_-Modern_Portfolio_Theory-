```python
# checking for null values
df_c.info()
```

```python
# selecting 365 days (1 year) 
df_c = df_c.head(365)
df_c
```

```python
# calculating % change for all the columns
# dropping null values(1st row)
# calculating daily mean return and annualised mean return
# calculating daily standard deviation and calculating annualised volatility
df_pct_change = df_c.pct_change()
df_pct_change = df_pct_change.dropna()
mean_pct_change = df_pct_change.mean()
std_pct_change = df_pct_change.std()
Annual_Volatility = std_pct_change * np.sqrt(252)
Annual_return = mean_pct_change * 252

print('Annual Volatilities for the 30 stocks are : \n', Annual_Volatility)
print('\n')
print('\n')
print('\n')
print('Annual Mean return for 30 stocks are :\n', Annual_return)
```

```python
# creating a data frame for Annual volatilities and Annual return
df_vr = pd.DataFrame(Annual_Volatility)
df_vr.rename(columns = {0: 'Annual_Volatilities'}, inplace = True)
df_vr['Annual_Return'] = Annual_return
df_vr
```

```python
# looping for elbow method to find the optimal value of K (total clusters)
# inertia is the sum of squared distances of samples to their closest cluster center.
# lower means better, but we choose where the graph drops suddenly and forms an elbow like shape
from sklearn.cluster import KMeans
Sum_of_squared_distances = []

for i in range(1,15):
    kmeans = KMeans(n_clusters = i)
    kmeans = kmeans.fit(df_vr)
    Sum_of_squared_distances.append(kmeans.inertia_)
```

```python
# we selected value of k = 5
# 'ro--' means red color marker of shape 'o' and line of type '--' (dashed)
plt.figure(figsize = (10,7))
plt.plot(range(1,15), Sum_of_squared_distances, 'ro--')
plt.xlabel('k')
plt.ylabel('Sum Of Squared Distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# applying the algo one last time to obtain results
kmeans = KMeans(n_clusters = 5)
kmeans.fit(df_vr)
```

```python
# obtaining the labels predicted by the algorithm
kmeans.labels_
```

```python
# plotting annual volatility vs annual returns 
# we set the color of clusters according to the labels (kmeans.labels_) and set the colour profile as rainbow
plt.plot(figsize = (10,7))
plt.scatter(x = df_vr.Annual_Return, y = df_vr.Annual_Volatilities, c = kmeans.labels_, cmap = 'rainbow' )
plt.title('K Means')
```

```python
# creating a data frame column named cluster
df_vr['Cluster'] = kmeans.labels_
df_vr = df_vr.drop(['Annual_Volatilities','Annual_Return'], axis = 1)
df_vr
```
